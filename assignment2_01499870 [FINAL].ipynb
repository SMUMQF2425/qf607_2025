#### Exercise 1.

Implement a trinomial tree that takes the below signature:

````python
def trinomialPricerAnchor(S, r, q, vol, trade, n, K):
    t = trade.expiry / n
    fwd = S * math.exp((r - q) * t * n)
    
    # Initial lambda to get close to K
    lmda = math.sqrt(3)
    u = math.exp(lmda * vol * math.sqrt(t))
    
    # Find node closest to K
    j_closest = int(round((math.log(fwd/K) / (lmda * vol * math.sqrt(t)) + n)))
    S_closest = S * u ** (n - j_closest)
    
    # Adjust lambda to exactly hit K
    if n - j_closest == 0:
        lmda_adj = lmda  # Use the original lambda if adjustment is not possible
    else:
        lmda_adj = math.log(fwd/K) / (vol * math.sqrt(t) * (n - j_closest))
    u = math.exp(lmda_adj * vol * math.sqrt(t))
    
    # Recompute probabilities with adjusted lambda
    mu = r - q
    pu = 1/(2*lmda_adj**2) + (mu - vol**2/2)/(2*lmda_adj*vol)*math.sqrt(t)
    pd = 1/(2*lmda_adj**2) - (mu - vol**2/2)/(2*lmda_adj*vol)*math.sqrt(t)
    pm = 1 - pu - pd
    
    # Tree construction
    vs = [trade.payoff(S * u ** (n - i)) for i in range(2*n + 1)]
    for i in range(n-1, -1, -1):
        new_vs = []
        for j in range(2*i + 1):
            continuation = math.exp(-r*t) * (pu*vs[j] + pm*vs[j+1] + pd*vs[j+2])
            nodeS = S * u ** (i - j)
            new_vs.append(trade.valueAtNode(t*i, nodeS, continuation))
        vs = new_vs
    return vs[0]
````

Instead of taking the parameter $\lambda$ in our example code, this function takes a discontinuity K and anchors a tree node to K by tuning $\lambda$ inside the pricer.
Note that there are many ways to adjust $\lambda$, one example is to construct the tree with $\lambda = \sqrt{3}$ first, find the node closest to $K$, then adjust $\lambda$ such that the node falls on $K$, but you are free to define your own way to achieve this.

Plot the pricing errors of this trinomial tree pricer with the market and strike below, for number of tree steps from 10 to 300 (similar to our binomial tree test cases).
Use a European Call option and a Binary Call option as test trades.

````python
K = 105
S, r, vol = 100, 0.01, 0.2

T = 1  # Assuming 1 year expiry
q = 0

# Analytical prices
def bs_price(S, K, T, r, vol, payoffType):
    d1 = (math.log(S/K) + (r + vol**2/2)*T) / (vol*math.sqrt(T))
    d2 = d1 - vol*math.sqrt(T)
    if payoffType == call:
        return S*norm.cdf(d1) - K*math.exp(-r*T)*norm.cdf(d2)
    else:  # Binary call
        return math.exp(-r*T) * norm.cdf(d2)

steps = range(10, 301, 10)
errors_call = []
errors_binary = []

class EuropeanCall:
    def __init__(self, expiry, strike):
        self.expiry = expiry
        self.strike = strike
    def payoff(self, S):
        return max(S - self.strike, 0)
    def valueAtNode(self, t, S, continuation):
        return continuation

class BinaryCall:
    def __init__(self, expiry, strike):
        self.expiry = expiry
        self.strike = strike
    def payoff(self, S):
        return 1.0 if S >= self.strike else 0.0
    def valueAtNode(self, t, S, continuation):
        return continuation

for n in steps:
    # Price European call
    call = EuropeanCall(T, K)
    price_tree = trinomialPricerAnchor(S, r, q, vol, call, n, K)
    price_bs = bs_price(S, K, T, r, vol, call)
    errors_call.append(abs(price_tree - price_bs))
    
    # Price binary call
    binary = BinaryCall(T, K)
    price_tree = trinomialPricerAnchor(S, r, q, vol, binary, n, K)
    price_bs = bs_price(S, K, T, r, vol, binary)
    errors_binary.append(abs(price_tree - price_bs))

plt.figure(figsize=(10,6))
plt.plot(steps, errors_call, label='European call error')
plt.plot(steps, errors_binary, label='Binary call error')
plt.xlabel('Number of steps')
plt.ylabel('Absolute error')
plt.legend()
plt.title('Trinomial tree pricing errors')
plt.show()
````

#### Exercise 2.

Assuming forward delta convention, implement below function ```strikeFromdelta``` that calculates the strike from a given delta.
You may use the ```fwdDelta``` function below. Test with below:

S, r, q, T, vol, delta, payoffType = 100, 0.1, 0, 1, 0.1, 0.25, PayoffType.Put

````python
class PayoffType(Enum):
    Call = 0
    Put = 1
def cnorm(x):
    return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0

def fwdDelta(fwd, stdev, strike, payoffType):
    d1 = math.log(fwd / strike) / stdev + stdev / 2
    if payoffType == PayoffType.Call:
        return cnorm(d1)
    elif payoffType == PayoffType.Put:
        return -cnorm(-d1)
    else:
        raise Exception("not supported payoff type", payoffType)

# solve for the K such that Delta(S, T, K, vol) = delta, assuming forward delta convention
def strikeFromDelta(S, r, q, T, vol, delta, payoffType):
    fwd = S * math.exp((r - q) * T)
    stdev = vol * math.sqrt(T)
    
    # For puts, the forward delta is negative by definition.
    target = delta if payoffType == PayoffType.Call else -delta
    
    def f(k):
        return fwdDelta(fwd, stdev, k, payoffType) - target
    
    # Initial bracket: you can adjust these bounds if necessary.
    a, b = 0.1 * S, 10 * S
    strike = optimize.brentq(f, a, b)
    return strike

S, r, q, T, vol, delta, payoffType = 100, 0.1, 0, 1, 0.1, 0.25, PayoffType.Put
strikeFromDelta(S, r, q, T, vol, delta, payoffType)
````

#### Exercise 3.

Implement ```smilefromMarks``` function below, it take the time to maturity $T$, current stop price $S$, risk free and dividend rate $r$ and $q$,
ATM volatility, BF25, RR25, BF10, RR10 quotes, returns a CubicSpline interpolated smile for this pillar. We assume ATM forward convention. 
You may use the Smile class provided below.   

````python
class Smile:
    def __init__(self, strikes, vols):
        # add additional point on the right to avoid arbitrage
        self.strikes = strikes  + [1.1*strikes[-1] - 0.1*strikes[-2]]
        self.vols = vols + [vols[-1] + (vols[-1]-vols[-2])/10]
        self.cs = CubicSpline(strikes, vols, bc_type=((1, 0.0), (1, 0.0)), extrapolate=True)
    def Vol(self, k):
        if k < self.strikes[0]:  # scipy cubicspline bc_type confusing, extrapolate by ourselfs
            return self.vols[0]
        if k > self.strikes[-1]:
            return self.vols[-1]
        else:
            return self.cs(k)

def smileFromMarks(T, S, r, q, atmvol, bf25, rr25, bf10, rr10):
    # Calculate strikes for 10P, 25P, ATM, 25C, 10C
    list_strikes = []
    list_vols = []
    
    # 10P
    k_10p = strikeFromDelta(S, r, q, T, atmvol + bf10 - rr10/2, 0.1, PayoffType.Put)
    list_strikes.append(k_10p)
    list_vols.append(atmvol + bf10 - rr10/2)
    
    # 25P
    k_25p = strikeFromDelta(S, r, q, T, atmvol + bf25 - rr25/2, 0.25, PayoffType.Put)
    list_strikes.append(k_25p)
    list_vols.append(atmvol + bf25 - rr25/2)
    
    # ATM (delta=0.5 for Call)
    atm_strike = S * math.exp((r - q) * T)
    list_strikes.append(atm_strike)
    list_vols.append(atmvol)
    
    # 25C
    k_25c = strikeFromDelta(S, r, q, T, atmvol + bf25 + rr25/2, 0.25, PayoffType.Call)
    list_strikes.append(k_25c)
    list_vols.append(atmvol + bf25 + rr25/2)
    
    # 10C
    k_10c = strikeFromDelta(S, r, q, T, atmvol + bf10 + rr10/2, 0.1, PayoffType.Call)
    list_strikes.append(k_10c)
    list_vols.append(atmvol + bf10 + rr10/2)
    
    # Sort strikes and vols and convert to lists (instead of tuples)
    strikes, vols = map(list, zip(*sorted(zip(list_strikes, list_vols))))
    
    return Smile(strikes, vols)
````

With ```smileFromMarks```, we can prepare our implied volatility surface from input marks.

````python
class ImpliedVol:
    def __init__(self, ts, smiles):
        self.ts = ts
        self.smiles = smiles
    # linear interpolation in variance, along the strike line
    def Vol(self, t, k):
        # locate the interval t is in
        pos = bisect.bisect_left(self.ts, t)
        # if t is on or in front of first pillar,
        if pos == 0:
            return self.smiles[0].Vol(k)
        if pos >= len(self.ts):
            return self.smiles[-1].Vol(k)
        else:  # in between two brackets
            prevVol, prevT = self.smiles[pos-1].Vol(k), self.ts[pos-1]
            nextVol, nextT = self.smiles[pos].Vol(k), self.ts[pos]
            w = (nextT - t) / (nextT - prevT)
            prevVar = prevVol * prevVol * prevT
            nextVar = nextVol * nextVol * nextT
            return  math.sqrt((w * prevVar + (1-w) * nextVar)/t)
        return
    def dVoldK(self, t, k):
        return (self.Vol(t, k+0.01) - self.Vol(t, k-0.01)) / 0.02
    def dVoldT(self, t, k):
        return (self.Vol(t+0.005, k) - self.Vol(t, k)) / 0.005
    def dVol2dK2(self, t, k):
        return (self.Vol(t, k+0.01) + self.Vol(t, k-0.01) - 2*self.Vol(t, k)) / 0.0001
````
